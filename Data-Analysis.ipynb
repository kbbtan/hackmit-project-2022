```python
import pandas as pd
import datetime

df = pd.read_csv("202208-bluebikes-tripdata.csv")
df.head()
```

```python
df['startDayNum'] = df['starttime'].dt.day_of_week
df['startDayName'] = df['starttime'].dt.day_name()
df.sort_values(by=["bikeid", "starttime"], inplace=True)
df.head()
```

```python
df['nextstarttime'] = df['starttime'].shift(-1)
df.head()
```
